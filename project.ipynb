```python
# Chi-Square Test for Categorical Columns
from scipy.stats import chi2_contingency
for col in ['Date','Teams','Venue','Toss_Winner','Toss_Decision','Win_Type','Player_of_Match','Umpire','Umpire1','Umpire2']:
    table=pd.crosstab(df1[col],df1['Match_Winner'],)
    _,p,_,_=chi2_contingency(table)
    print(f"{col}:p-vlaue={p:.4f}->{'significant' if p<0.05 else 'not significant'}")
```

```python

for col in df1.select_dtypes(include='number').columns:
    Q1 = df1[col].quantile(0.25)
    Q3 = df1[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df1 = df1[(df1[col] >= lower) & (df1[col] <= upper)]
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df1['Player_of_Match'] = le.fit_transform(df1['Player_of_Match'])
df1 = pd.get_dummies(df1, columns=[  
    'Teams', 
    'Venue', 
    'Toss_Winner', 
    'Toss_Decision'
], drop_first=True)
```

```python
df1.dtypes
```

```python
x=df1.drop(['Match_Winner'],axis=1).values
x
```

```python
y=df1['Match_Winner'].values
y
```

```python
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler(sampling_strategy='auto', random_state=42)
X_resampled, y_resampled = rus.fit_resample(x,y)
x=X_resampled
y=y_resampled
```

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42) 
x_train 
```

```python
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler() 
scaler.fit(x_train) 
x_train=scaler.transform(x_train) 
x_test=scaler.transform(x_test) 
```

```python
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.naive_bayes import BernoulliNB 
from sklearn.svm import SVC 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.ensemble import RandomForestClassifier 
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
knn=KNeighborsClassifier(n_neighbors=7) 
naive=BernoulliNB() 
support=SVC() 
tree=DecisionTreeClassifier() 
forest=RandomForestClassifier(n_estimators=100) 
lst=[knn,naive,support,tree,forest]
```

```python
for i in lst: 
    i.fit(x_train,y_train) 
    y_predict=i.predict(x_test) 
    print("performance evaluation of :- ",i)  
    print("*"*100) 
    print("accuracy score is :-",accuracy_score(y_test,y_predict)) 
    print("*"*100) 
    print("classifiaction report is :-",classification_report(y_test,y_predict)) 
    print("*"*100)    
```
